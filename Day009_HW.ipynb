取得 SIFT 特徵

- 轉成灰階圖片
- 需要額外安裝 OpenCV 相關套件

```python
import numpy as np
import cv2
import copy
from google.colab.patches import cv2_imshow

# !wget http://cvdl.cupoy.com/file-download/part01/lena.png -O lena.png
img = cv2.imread('./lena.png')
```

```python
# 轉灰階圖片
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 建立 SIFT 物件
SIFT_detector = cv2.xfeatures2d.SIFT_create()

# 取得 SIFT 關鍵點位置
keypoints = SIFT_detector.detect(img_gray, None)

#　畫圖 + 顯示圖片
img_show = cv2.drawKeypoints(img_gray, keypoints, img)
cv2_imshow(img_show)

# while True:
#     cv2.imshow('SIFT', img_show)
#     k = cv2.waitKey(0)
#     if k == 27:
#         cv2.destroyAllWindows()
#         break
```

```python
# !wget http://cvdl.cupoy.com/file-download/part01/lena.png -O lena1.png

img1 = cv2.imread('./lena1.png')

def sift (img,img_gray):
  # 建立 SIFT 物件
  SIFT_detector = cv2.xfeatures2d.SIFT_create()

  # 取得 SIFT 關鍵點位置
  keypoints = SIFT_detector.detect(img_gray, None)

  #　畫圖 + 顯示圖片
  img_show = cv2.drawKeypoints(img_gray, keypoints, img)
  cv2_imshow(img_show)
  

#顯示圖片
cv2_imshow(img1)

# 顯示彩圖(R)
imgR = copy.copy(img1)
imgR[:,:,0] = 0
imgR[:,:,1] = 0


# 顯示彩圖(G)
imgG = copy.copy(img1)
imgG[:,:,0] = 0
imgG[:,:,2] = 0


# 顯示彩圖(B)
imgB = copy.copy(img1)
imgB[:,:,1] = 0
imgB[:,:,2] = 0

sift (img1,imgR)
sift (img1,imgG)
sift (img1,imgB)
```
