# Import

```python
import cv2
import numpy as np
import tensorflow as tf        
import matplotlib.pyplot as plt                        
%matplotlib inline
```

# 載入圖片

```python
!wget "https://upload.wikimedia.org/wikipedia/commons/thumb/2/2a/FIRE_01.JPG/300px-FIRE_01.JPG" -O dt.jpg
```

# 讀取圖片

```python
image = cv2.imread('./dt.jpg')

image.shape
```

#  顯示圖片

```python
# 使用 OpenCV 讀取圖檔
plt.figure(figsize=(9,4))

plt.subplot(121)
plt.imshow(image)

# 將 BGR 圖片轉為 RGB 圖片
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#img_rgb = image[:,:,::-1]
plt.subplot(122)
plt.imshow(img_rgb)
plt.show()
```

# 顯示channel圖片

```python
plt.figure(figsize=(14,6))

# 顯示彩圖(R)
imgR = image.copy()
imgR[:,:,1] = 0
imgR[:,:,2] = 0

plt.subplot(131)
plt.imshow(imgR)

# 顯示彩圖(G)
imgG = image.copy()
imgG[:,:,0] = 0
imgG[:,:,2] = 0

plt.subplot(132)
plt.imshow(imgG)

# 顯示彩圖(B)
imgB = image.copy()
imgB[:,:,0] = 0
imgB[:,:,1] = 0
plt.subplot(133)
plt.imshow(imgB)

plt.show()
```
