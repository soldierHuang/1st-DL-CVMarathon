# Import

```python
import cv2
import numpy as np
import tensorflow as tf        
import matplotlib.pyplot as plt                        
%matplotlib inline
```

# 載入圖片

```python
!wget "http://optipng.sourceforge.net/pngtech/img/lena.png" -O Lena.jpg
```

# 讀取圖片

```python
img = cv2.imread('./Lena.jpg', cv2.IMREAD_COLOR)
img.shape
```

# 改變不同的 color space

```python
# 將 BGR 圖片轉為 RGB 圖片
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#img_RGB = image[:,:,::-1]

# HSV
img_HSV = cv2.cvtColor(img_RGB, cv2.COLOR_BGR2HSV)

# HLS
img_HLS = cv2.cvtColor(img_RGB, cv2.COLOR_BGR2HLS)

# LAB
img_LAB = cv2.cvtColor(img_RGB, cv2.COLOR_BGR2LAB)
```

# 顯示圖片

```python
plt.figure(figsize=(18,7))

plt.subplot(151)
plt.imshow(img)

plt.subplot(152)
plt.imshow(img_RGB)

plt.subplot(153)
plt.imshow(img_HSV)

plt.subplot(154)
plt.imshow(img_HLS)

plt.subplot(155)
plt.imshow(img_LAB)

plt.show()
```
